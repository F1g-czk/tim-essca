## Étape 2 - Affichage d'un tableau



```java
// On utilise le + pour assembler le texte et les valeurs
System.out.println("Produit" + "  | " + "PU" + "    | " + "Qté" + " | " + "Prix");
System.out.println("---------------------------------");
System.out.println("Stylos" + "   | " + 1.20 + "  | " + 5 + "   | " + (1.20 * 5));
System.out.println("Cahiers" + "  | " + 3.50 + "  | " + 2 + "   | " + (3.50 * 2));
System.out.println("---------------------------------");
System.out.println("TOTAL" + "                  | " + (1.20 * 5 + 3.50 * 2));
```

```java
// Affichage du tableau 
System.out.println("+------------+-------+-----+-------+");
System.out.println("| " + "Produit" + "    | " + "PU" + "    | " + "Qté" + " | " + "Prix" + "  |");
System.out.println("+------------+-------+-----+-------+");

// Ligne pour les stylos
System.out.println("| " + "Stylos" + "     | " + 1.20 + "  | " + 5 + "   | " + (1.20 * 5) + "  |");

// Ligne pour les cahiers
System.out.println("| " + "Cahiers" + "    | " + 3.50 + "  | " + 2 + "   | " + (3.50 * 2) + "  |");

System.out.println("+------------+-------+-----+-------+");

// Ligne du total
System.out.println("| TOTAL" + "                    | " + (1.20 * 5 + 3.50 * 2) + " |");
System.out.println("+------------+-------+-----+-------+");
```

## Étape 3 - Affichage de variables

```java
// variables
String nomProduit1 = "Stylos";
double prixUnitaire1 = 1.20;
int quantite1 = 5;

String nomProduit2 = "Cahiers";
double prixUnitaire2 = 3.50;
int quantite2 = 2;

// Calcul du total 
double total = (prixUnitaire1 * quantite1) + (prixUnitaire2 * quantite2);

// tableay
System.out.println("+------------+-------+-----+-------+");
System.out.println("| " + "Produit" + "    | " + "PU" + "    | " + "Qté" + " | " + "Prix" + "  |");
System.out.println("+------------+-------+-----+-------+");

// Utilisation de printf 
System.out.printf("| %-10s | %5.2f | %3d | %5.2f |\n", nomProduit1, prixUnitaire1, quantite1, (prixUnitaire1 * quantite1));
System.out.printf("| %-10s | %5.2f | %3d | %5.2f |\n", nomProduit2, prixUnitaire2, quantite2, (prixUnitaire2 * quantite2));

System.out.println("+------------+-------+-----+-------+");
System.out.printf("| TOTAL GENERAL            | %5.2f |\n", total);
System.out.println("+------------+-------+-----+-------+");
```

## Étape 4 - Saisie utilisateur

```java
import java.util.Scanner;

Scanner clavier = new Scanner(System.in);

// 1. Saisie des info
System.out.print("Entrez la quantité de " + nomProduit1 + " : ");
quantite1 = clavier.nextInt();

System.out.print("Entrez la quantité de " + nomProduit2 + " : ");
quantite2 = clavier.nextInt();

// 2. Calcul destotaux
double total1 = prixUnitaire1 * quantite1;
double total2 = prixUnitaire2 * quantite2;
double totalGeneral = total1 + total2;

// 3. Affichage du tableaux avec les caractères +, - et |
System.out.println("\n+------------+-------+-----+-------+");
System.out.println("| Produit    | PU    | Qté | Prix  |");
System.out.println("+------------+-------+-----+-------+");

// Utilisation de printf pour que les données restent bien alignées dans les cases
System.out.printf("| %-10s | %5.2f | %3d | %5.2f |\n", nomProduit1, prixUnitaire1, quantite1, total1);
System.out.printf("| %-10s | %5.2f | %3d | %5.2f |\n", nomProduit2, prixUnitaire2, quantite2, total2);

System.out.println("+------------+-------+-----+-------+");
System.out.printf("| TOTAL GENERAL            | %5.2f |\n", totalGeneral);
System.out.println("+------------+-------+-----+-------+");
```
